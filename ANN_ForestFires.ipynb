```python
forest1= forest.iloc[:,2:30]
```

```python
forest1.head()
```

**Standardization**

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
SC.fit(forest1)
forest_norm = SC.transform(forest1)
forest_norm
```

**Used Standard Scalar in order to Normalized the data to make it easier to compare and analyze.**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 28)
pca_values = pca.fit_transform(forest_norm)
pca_values
```

**Applied PCA on standardized dataset,to reduce dimensionality of the dataset.** 

**After applying PCA retaining Varinaces for 28 components**

```python
var = pca.explained_variance_ratio_
var
```

**Obtaining cumulative sum of the Variance that we got in the previous step for each component of PCA.**

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
final = pd.concat([pd.DataFrame(pca_values[:,0:28],columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7',
 'pc8','pc9','pc10','pc11','pc12','pc13','pc14',
 'pc15','pc16','pc17','pc18','pc19','pc20','pc21',
 'pc22','pc23','pc24','pc25','pc26','pc27','pc28']),
 forest[['size_category']]], axis = 1)
final.size_category.replace(('large','small'),(1,0),inplace=True)
final
```

```python
import seaborn as sns
fig=plt.figure(figsize=(20,10))
sns.scatterplot(data=final)
```

**This scatter plot is used to visualize the relationship between the principal components and the size_category variable.**

```python
array = final.values
X = array[:,0:28]
Y = array[:,28]
```

```python
X
```

```python
Y
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10, input_dim=28, activation='relu'))
model.add(Dense(6,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

**Making model to learn the pattern with input data and predict the corresponding binary labels with good accuracy.**

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, validation_split=0.3, epochs=150, batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

**Got Accuracy Score as 93% which means that our model is 93% accurately predicting the output results as 1 or 0.**
