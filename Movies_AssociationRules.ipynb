```python
#Importing Libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import warnings
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv("/content/my_movies.csv")
movies.head()
```

```python
movies.columns
```

```python
movies = movies[['SixthSense', 'Gladiator', 'LOTR1','HarryPotter1', 'Patriot', 'LOTR2', 'HarryPotter2', 'LOTR','Braveheart', 'GreenMile']]
```

```python
movies.head()
```

```python
for i in movies.columns:
    print(i)
    print(movies[i].value_counts())
    print()
```

# Apriori Algorithm

```python
frequent_itemsets = apriori(movies, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
movies[movies["Gladiator"]==1].shape[0]/movies.shape[0]
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
rules.sort_values('lift',ascending = False)
```

```python
rules = association_rules(frequent_itemsets,metric = "lift", min_threshold=1.0)
rules
rules.sort_values('lift',ascending = False)
```

An leverage value of 0 indicates independence. Range will be [-1 1]
A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold=1.4)
rules[rules["lift"]>1]
```
