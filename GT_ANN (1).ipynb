```python
def select_features(X_train, Y_train, X_test):
    fs = SelectKBest(score_func=mutual_info_regression, k='all')
    fs.fit(X_train, Y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
```

```python
#Splitting the dataset using train & test split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.33, random_state=1)
X_train_fs, X_test_fs, fs = select_features(X_train, Y_train, X_test)
```

```python
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
fig, ax = plt.subplots(figsize=(15, 8))
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

**We created bar plot to visualize the mutual information scores of each feature,these are the most important features for predicting gas turbine energy yield using mutual information regression.**







**Looking at this visualization we can say that AT,AP,AH, CO & NOX these scored very less or we can say that these features are less important so we can drop them as they contribute very less.**

# Creating Model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def create_model(learning_rate,dropout_rate,activation_function,init,neuron1,neuron2):
    model = Sequential()
    model.add(Dense(neuron1,input_dim = 5,kernel_initializer = init,activation = activation_function))
    model.add(Dropout(dropout_rate))
    model.add(Dense(neuron2,input_dim = neuron1,kernel_initializer = init,activation = activation_function))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1,activation = 'linear'))
    
    adam=adam_v2.Adam(learning_rate = learning_rate)
    model.compile(loss = 'mean_squared_error',optimizer = adam,metrics = ['mse'])
    return model
```

```python
# Creating model
model = KerasClassifier(build_fn = create_model,verbose = 0)
# Defining grid search parameters
batch_size = [20,40]
epochs = [50,100]
learning_rate = [0.01,0.1]
dropout_rate = [0.1,0.2]
activation_function = ['relu','linear']
init = ['uniform','normal']
neuron1 = [4,8]
neuron2 = [2,4]
# Make dictionary of the grid search parameters
param_grids = dict(batch_size = batch_size,epochs = epochs,learning_rate = learning_rate,dropout_rate = dropout_rate,activation_function = activation_function,init = init,neuron1 = neuron1,neuron2 = neuron2)
# Build and fit the GridSearchCV
grid = GridSearchCV(estimator = model,param_grid = param_grids,cv = KFold(),verbose = 10, scoring='neg_mean_squared_error')
grid_result = grid.fit(x_train, y_train)
#results
print('Best : {}, using {}'.format(grid_result.best_score_,grid_result.best_params_))
```

```python
final_model = Sequential()
final_model.add(Dense(4,input_dim = 5,kernel_initializer = 'uniform',activation = 'relu'))
final_model.add(Dropout(0.1))
final_model.add(Dense(2,input_dim = 4,kernel_initializer = 'uniform',activation = 'relu'))
final_model.add(Dropout(0.1))
final_model.add(Dense(1,activation = 'linear'))
    
adam=adam_v2.Adam(learning_rate = 0.01)
```

```python
final_model.compile(loss = 'mean_squared_error',optimizer = adam,metrics = ['mse'])
```

```python
final_model.fit(x_train,y_train, epochs=50, batch_size=20)
```

```python
final_model.summary()
```

```python
#evaluating the model
scores = final_model.evaluate(x_test, y_test)
print((final_model.metrics_names[1]))
```
